```python
import pandas as pd
from flashtext import KeywordProcessor
```

```python
df=pd.read_csv("Blackbox.csv")
```

```python
df
```

```python
tastelist = df.Taste.to_list()
```

```python
tastelist
```

```python
tastelist = [x for x in tastelist if x == x]
```

```python
tastelist
```

```python
Flavorlist = df.Flavor.to_list()
Flavorlist = [x for x in Flavorlist if x == x]
Flavorlist
```

```python
Texturelist = df.Texture.to_list()
Texturelist = [x for x in Texturelist if x == x]
Texturelist
```

```python
Occasionlist = df.Occasion.to_list()
Occasionlist = [x for x in Occasionlist if x == x]
Occasionlist
```

```python
Companylist = df.Company.to_list()
Companylist = [x for x in Companylist if x == x]
Companylist
```

```python
df.columns
```

```python
tlist = KeywordProcessor()
tlist.add_keywords_from_list(tastelist)
```

```python
df['Taste.1'] = [tlist.extract_keywords(x) for x in df['combined taste/flavor/texture/appearance']]
```

```python
df['Taste.1']
```

```python
Flist = KeywordProcessor()
Flist.add_keywords_from_list(Flavorlist)
Tlist = KeywordProcessor()
Tlist.add_keywords_from_list(Texturelist)
olist = KeywordProcessor()
olist.add_keywords_from_list(Occasionlist)
colist = KeywordProcessor()
colist.add_keywords_from_list(Companylist)
```

```python
df.columns
```

```python
df['Flavor.1'] = [Flist.extract_keywords(x) for x in df['combined taste/flavor/texture/appearance']]
df['Texture.1'] = [Tlist.extract_keywords(x) for x in df['combined taste/flavor/texture/appearance']]
df['Occasion.1'] = [olist.extract_keywords(x) for x in df['combined occasion/taste']]
df['Company.1'] = [colist.extract_keywords(x) for x in df['combined occasion/taste']]
```
